```python
import tkinter as tk
from tkinter import messagebox
import random

class TroChoiDoanChu:
    def __init__(self):
        try:
            with open("words.txt", "r", encoding="utf-8") as f:
                self.ds_tu = [dong.strip().upper() for dong in f.readlines()]
            with open("cauhoi.txt", "r", encoding="utf-8") as f:
                self.ds_goi_y = [dong.strip() for dong in f.readlines()]

            if len(self.ds_tu) != len(self.ds_goi_y):
                raise ValueError("Số lượng từ và gợi ý không khớp!")
        except Exception as loi:
            messagebox.showerror("Lỗi", f"Lỗi khi đọc file: {loi}")
            exit()

        self.khoi_dong_tu_moi()

    def khoi_dong_tu_moi(self):
        chi_so = random.randint(0, len(self.ds_tu) - 1)
        self.tu_dap_an = self.ds_tu[chi_so]
        self.goi_y = self.ds_goi_y[chi_so]
        self.cac_chu_da_doan = set()
        self.so_lan_doan_sai = 0

    def doan_chu(self, chu):
        if chu in self.cac_chu_da_doan:
            return False
        self.cac_chu_da_doan.add(chu)
        if chu not in self.tu_dap_an:
            self.so_lan_doan_sai += 1
            return False
        return True

    def hien_thi_tu(self):
        return " ".join([chu if chu in self.cac_chu_da_doan else "_" for chu in self.tu_dap_an])

    def da_thang(self):
        return all(chu in self.cac_chu_da_doan for chu in self.tu_dap_an)

    def da_thua(self):
        return self.so_lan_doan_sai >= 6

class GiaoDienDoanChu:
    def __init__(self, root):
        self.root = root
        self.root.title("Trò chơi Đoán Chữ")
        self.tro_choi = TroChoiDoanChu()

        self.nhan_tu = tk.Label(root, text="", font=("Helvetica", 20))
        self.nhan_tu.pack(pady=10)

        self.nhan_goi_y = tk.Label(root, text="", font=("Helvetica", 14), fg="blue")
        self.nhan_goi_y.pack(pady=5)

        self.khung_nhap = tk.Frame(root)
        self.khung_nhap.pack()

        self.o_nhap_chu = tk.Entry(self.khung_nhap, font=("Helvetica", 16), width=5)
        self.o_nhap_chu.grid(row=0, column=0, padx=5)

        self.nut_doan = tk.Button(self.khung_nhap, text="Đoán", command=self.xu_ly_doan)
        self.nut_doan.grid(row=0, column=1, padx=5)

        self.nhan_sai = tk.Label(root, text="Sai: 0", font=("Helvetica", 14), fg="red")
        self.nhan_sai.pack()

        self.bang_ve = tk.Canvas(root, width=200, height=250, bg="white")
        self.bang_ve.pack(pady=10)

        self.nut_choi_lai = tk.Button(root, text="Chơi lại", command=self.khoi_dong_lai)
        self.nut_choi_lai.pack()

        self.root.bind('<Return>', lambda su_kien: self.xu_ly_doan())

        self.khoi_dong_lai()

    def khoi_dong_lai(self):
        self.tro_choi.khoi_dong_tu_moi()
        self.bang_ve.delete("all")
        self.nhan_tu.config(text=self.tro_choi.hien_thi_tu())
        self.nhan_goi_y.config(text=f"Gợi ý: {self.tro_choi.goi_y}")
        self.nhan_sai.config(text="Sai: 0")

        # Vẽ giá treo
        self.bang_ve.create_line(20, 230, 180, 230)  
        self.bang_ve.create_line(50, 230, 50, 20)    
        self.bang_ve.create_line(50, 20, 130, 20)    
        self.bang_ve.create_line(130, 20, 130, 40)   

    def xu_ly_doan(self):
        chu = self.o_nhap_chu.get().strip().upper()
        self.o_nhap_chu.delete(0, tk.END)

        if len(chu) != 1 or not chu.isalpha():
            messagebox.showwarning("Cảnh báo", f"Vui lòng nhập 1 chữ cái. Bạn đã nhập: '{chu}'")
            return

        self.tro_choi.doan_chu(chu)
        self.nhan_tu.config(text=self.tro_choi.hien_thi_tu())
        self.nhan_sai.config(text=f"Sai: {self.tro_choi.so_lan_doan_sai}")
        self.ve_nguoi(self.tro_choi.so_lan_doan_sai)

        if self.tro_choi.da_thang():
            messagebox.showinfo("Chiến thắng", "Bạn đã đoán đúng từ!")
            self.khoi_dong_lai()
        elif self.tro_choi.da_thua():
            messagebox.showinfo("Thua cuộc", f"Bạn đã thua! Từ đúng là: {self.tro_choi.tu_dap_an}")
            self.khoi_dong_lai()

    def ve_nguoi(self, so_sai):
        if so_sai >= 1:
            self.bang_ve.create_oval(110, 40, 150, 80)        
        if so_sai >= 2:
            self.bang_ve.create_line(130, 80, 130, 140)       
        if so_sai >= 3:
            self.bang_ve.create_line(130, 100, 100, 120)      
        if so_sai >= 4:
            self.bang_ve.create_line(130, 100, 160, 120)      
        if so_sai >= 5:
            self.bang_ve.create_line(130, 140, 110, 180)      
        if so_sai >= 6:
            self.bang_ve.create_line(130, 140, 150, 180)      

# ---------------- BẮT ĐẦU CHẠY GAME ---------------- #
if __name__ == "__main__":
    root = tk.Tk()
    ung_dung = GiaoDienDoanChu(root)
    root.mainloop()
```
